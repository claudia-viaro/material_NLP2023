## Prepare data

```python
from datasets import load_dataset

train_data = load_dataset("imdb",split='train')
test_data = load_dataset("imdb",split='test')
splitted_data = train_data.train_test_split(test_size=0.1)

train_df = splitted_data['train'].to_pandas()[:1000]
eval_df = splitted_data['test'].to_pandas()[:200]
test_df = test_data.to_pandas()[:200]

train_df.columns = ["text", "labels"]
eval_df.columns = ["text", "labels"]
test_df.columns = ["text", "labels"]

print(train_df)
print(f'Training samples = {len(train_df)} | Validation samples = {len(eval_df)} | Test samples = {len(test_df)}')
```

## Training Model

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd

# Optional model configuration
model_args = ClassificationArgs(num_train_epochs=1)
model_args.learning_rate = 1e-4
model_args.evaluate_during_training = True
model_args.overwrite_output_dir = True
model_args.train_batch_size = 8
model_args.wandb_project = 'RANLP_1'
model_args.num_train_epochs = 1

# Create a ClassificationModel
model = ClassificationModel(
    'bert',
    'bert-base-uncased',
    args=model_args
)

# Train the model
model.train_model(train_df,eval_df=eval_df)
```

```python

# Evaluate The model
predictions, raw_outputs, wrong = model.eval_model(test_df)
```
